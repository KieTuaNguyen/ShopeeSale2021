## Data cleaning and wrangling steps

### Check missing values

```python
print(raw_df.isnull().sum()/len(raw_df)*100)
```

As we can see, there are around 12,5 % missing values in the `product_price` column. 
Then, we will plot the distribution of the `product_price` column to see if we can replace the missing values with the mean or median.

```python
raw_df['product_price'].plot(kind='hist', figsize=(8, 5))
```

As we can see, there are under 40% of the `product_price` values. 
Therefore, we will replace the missing values with the highest frequency value.

### Handle Missing Values

We will replace the missing values in `product_price` with the highest frequency value.

```python
raw_df['product_price'].fillna(raw_df['product_price'].mode()[0], inplace=True)
```

We will double check if there are any missing values left.

```python
print(raw_df.isnull().sum()/len(raw_df)*100)
```
